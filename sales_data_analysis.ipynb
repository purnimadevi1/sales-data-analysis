```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("superstore_sales.csv")
df.head()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst=True)
```

```python
df['Month'] = df['Order Date'].dt.month
monthly_sales = df.groupby('Month')['Sales'].sum()
monthly_sales
```

```python
region_sales = df.groupby('Region')['Sales'].sum()
region_sales
```

```python
region_sales.plot(kind='bar')
plt.title("Sales by Region")
plt.xlabel("Region")
plt.ylabel("Revenue")
plt.show()
```

```python
category_sales = df.groupby('Category')['Sales'].sum()
category_sales
```

```python
category_sales.plot(kind='bar')
plt.title("Sales by Category")
plt.xlabel("Category")
plt.ylabel("Revenue")
plt.show()
```

```python
top_products = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False)
top_products.head(10)
```

```python
top_cities = df.groupby('City')['Sales'].sum().sort_values(ascending=False)
top_cities.head(10)
```

### Insights

- The West region generates the highest revenue.
- Technology category contributes the most sales.
- A small number of products dominate overall revenue.
- Certain cities contribute significantly to total sales.
